```python
import pandas as pd
```

***Load the dataset -
We start by loading the raw Netflix dataset and inspecting its structure***

```python
df=pd.read_csv("/content/netflix1.csv.zip")
df
```

```python
df.head()
```

```python
df.info()
```

```python
df.columns
```

***Deleting redundant columns***

```python
df=df.drop(columns="rating")

```

```python
df
```

***Renaming the columns***

```python
new_column_names=[]

for i in df.columns:
  new_column_names.append(i.capitalize())
```

```python
new_column_names
```

```python
df.columns = new_column_names
```

```python
df.head()
```

***Handling dulpicates***

```python
df.duplicated().sum()
```

*in the above case there is no duplictaes, but in case of duplictaes, we can drop the duplicate row using:
df.drop_duplicates(inplace = true)*

***Checking missing values***

```python
df.isna().sum()
```

*If we had na values : df.dropna(inplace = true)*

***Check for more transformation***

```python
df["Show_id"].unique()
```

```python
df["Show_id"] = df["Show_id"].apply(lambda x: x.split("s")[1])
```

```python
df["Show_id"]
```

```python
type(df["Show_id"][0])
```

*Change the type from String to integer*

```python
df["Show_id"] = df["Show_id"].astype(int)
```

```python
type(df["Show_id"][0])
```

```python
df["Date_added"] = df["Date_added"].apply(lambda x: x.replace("/","-"))
```

```python
df["Date_added"] = pd.to_datetime(df["Date_added"])
```

```python
df
```

```python
df.to_csv("netflix_cleaned.csv", index=False)
```
